# Your mission is to determine which date it was / it will be

# Votre mission est de déterminer quelle date c'était / ce sera

```
# what month will it be in 7 weeks ?
# Quel mois ce sera dans 7 semaines ?

var0 = datetime.now() + timedelta(weeks=7)
month_name = var0.strftime("%B")
print(datetime.now() + timedelta(weeks=7))
print(month_name)

# Autrement:
# print((datetime.now() + timedelta(weeks=7)).strftime('%B'))
```

```
# what day of the week will it be in 3 days ?
# Quel jour de la semaine sera-t-il dans 3 jours ?

var11 = datetime.now() + timedelta(days=3)
day_name = var0.strftime("%A")
print(datetime.now() + timedelta(days=3))
print(day_name)

#Autrement:
#print((datetime.now() + timedelta(days=3)).strftime('%A'))
```

```
# what hour will it be in 100 hours ?
# Quelle heure sera-t-il dans 100 heures ?

print(datetime.now() + timedelta(hours=100))

#Autrement
#print((datetime.now() + timedelta(hours=100)).strftime('%H:%M'))
```

```
# what date was 30 days ago ?
# Quelle date était il y a 30 jours ?

print(datetime.now() - timedelta(days=30))

#Autrement
#print((datetime.now() - timedelta(days=30)).date().isoformat())
```

# With Pandas
"DATE" column has only string values. On a chart, the X-axis is not well displayed.

```
# Please run this cell without modification:
df = pd.read_csv("https://raw.githubusercontent.com/murpi/wilddata/master/quests/weather2019.csv")
fig, ax = plt.subplots(figsize=(20, 10))
ax.scatter(x=df['DATE'],
           y=df['MAX_TEMPERATURE_C'])
fig.show()
```

```
# Convert the "DATE" column into a datetime format, then display the same chart. Are ticks on X-axis better printed ?

# Convertir la colonne "DATE" en un format de date, puis afficher le même graphique. Les tics sur l'axe des X sont-ils mieux imprimés ?
# Réponse : oui, autant mieux que la cellule dessous, sinon un peu petit peu éclaté !

df['DATE'] = df['DATE'].apply(lambda x: datetime.strptime(x, '%Y-%m-%d'))

fig1, ax = plt.subplots(figsize=(20, 10))
ax.scatter(x=df['DATE'],
           y=df['MAX_TEMPERATURE_C'])

fig1.show()
```

```
# Afficher la table:

df['DATE'] = pd.to_datetime(df['DATE'])
df.head()
```

```
# Display only rows for the month of June
# Afficher uniquement les lignes du mois de juin

df = df[pd.to_datetime(df['DATE']).dt.month == 6]

fig2, ax = plt.subplots(figsize=(20, 10))
ax.scatter(x=df['DATE'],
           y=df['MAX_TEMPERATURE_C'])

fig2.show()
```
