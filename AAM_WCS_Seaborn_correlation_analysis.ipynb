## Pair Plot - Coloriser les opinions
Une autre façon, plus visuelle, de voir la corrélation, consiste à représenter chaque colonne 2 par 2, c'est le pairplot. Mais il est lent à calculer sur l'ensemble de la base de données.

Crée une tranche de 10 colonnes (y compris la colonne "OPINION")
Crée un échantillon de cette tranche
Dessine un diagramme de paires sur cet échantillon, et le colore en fonction de l'`OPINION`.

```
# It's your turn:
# C'est votre tour maintenant :

#Crée une tranche de 10 colonnes (y compris la colonne "OPINION")Crée un échantillon de cette tranche
data = df_weather.loc[:, 'DATE': 'HUMIDITY_MAX_PERCENT']
slice10 = pd.concat([data, df_weather['OPINION']], axis=1)
slice10
```

```
# Dessine un diagramme de paires sur cet échantillon, et le colore en fonction de l'OPINION.
sns.pairplot(slice10, hue='OPINION')
```

Draw 4 different charts of your choice on the same dashboard. You're an analyst, you have to be force of proposal. Your goal here is to represent which season has the best weather.

# De plus en plus...
Dessinez 4 graphiques différents de votre choix sur le même tableau de bord. Vous êtes analyste, vous devez être force de proposition. Votre objectif ici est de représenter quelle saison a le meilleur temps.

N'oubliez pas qu'une "figure" MatPlotLib est comme un tableau de bord incluant des sous-plots. Vous devez donc créer une figure MatPlotLib avec des sous-plots. Et ces sous-plots seront les graphiques de Seaborn.

```
# Your code here :

# votre code ici :

fig, axes = plt.subplots(2, 2, figsize=(18, 12))
fig.suptitle('Seasons By Opinion')

# saison hiver
df_weather1 = df_weather.loc[(df_weather['DATE'].dt.month==12) | (df_weather['DATE'].dt.month==1) | (df_weather['DATE'].dt.month==2)]
sns.scatterplot(ax=axes[0, 0], data= df_weather1, x= 'DATE', y='OPINION', hue ='OPINION')
plt.legend(loc='best')
plt.xlabel("DATE")
axes[0,0].set_title('WINTER')


# Saison de primtemps
df_weather2 = df_weather.loc[(df_weather['DATE'].dt.month==3) | (df_weather['DATE'].dt.month==4) | (df_weather['DATE'].dt.month==5)]
sns.scatterplot(ax=axes[0, 1], data= df_weather2, x= 'DATE', y='OPINION', hue='OPINION')
plt.legend(loc='best')
plt.xlabel("DATE")
axes[0,1].set_title("SPRING")


# Saison de l'été
df_weather3 = df_weather.loc[(df_weather['DATE'].dt.month==6) | (df_weather['DATE'].dt.month==7) | (df_weather['DATE'].dt.month==8)]
sns.scatterplot(ax=axes[1, 0], data= df_weather3, x= 'DATE', y='OPINION', hue='OPINION')
plt.legend(loc='best')
plt.xlabel("DATE")
axes[1,0].set_title("SUMMER")


# Season Automne
df_weather4 = df_weather.loc[(df_weather['DATE'].dt.month==9) | (df_weather['DATE'].dt.month==10) | (df_weather['DATE'].dt.month==11)]
sns.scatterplot(ax=axes[1, 1], data= df_weather4, x= 'DATE', y='OPINION', hue='OPINION')
plt.legend(loc='best')
plt.xlabel("DATE")
axes[1,1].set_title("AUTUMN")


plt.show()
```
