# PIVOT TABLE

## It's like in Excel, but with Pandas

There are 2 syntaxes, choose your preferred :
- As a method : YOUR_DATAFRAME.pivot_table()
- As a function : pd.crosstab()

## C'est comme dans Excel, mais avec Pandas

Il y a 2 syntaxes, choisissez celle que vous préférez :
- En tant que méthode : YOUR_DATAFRAME.pivot_table()
- Comme une fonction : pd.crosstab()

```
# Example
df_weather.pivot_table(values = 'MAX_TEMPERATURE_C', index = 'MONTH', aggfunc = 'mean')
```

```
# Autre synthaxe: Méthode crosstab():
pd.crosstab(values = df_weather['MAX_TEMPERATURE_C'],
            index = df_weather['MONTH'], columns ="",
            aggfunc = 'mean')
```

```
# Autrement:

# Your code here
# Votre code ici:
import matplotlib.pyplot as plt
# First I reset the index to buid my chart based on this datafram
df_max_temp_sunhour = pd.pivot_table(df_weather,index=["MONTH"],values=["MAX_TEMPERATURE_C","SUNHOUR"],
               aggfunc={"MAX_TEMPERATURE_C":np.mean,"SUNHOUR":np.mean},fill_value=0)
df_max_temp_sunhour_db = df_max_temp_sunhour.reset_index()


plt.title("Temperatures et durée d'ensoleillement")
plt.ylabel("")
x = df_max_temp_sunhour_db["MONTH"]
y = df_max_temp_sunhour_db["MAX_TEMPERATURE_C"]
plt.bar(x, y, label = 'Temp max')
x = df_max_temp_sunhour_db["MONTH"]
y = df_max_temp_sunhour_db["SUNHOUR"]
plt.plot(x, y, c= 'red', label = 'Ensoleillemnt')
plt.legend(loc="upper left")
```
