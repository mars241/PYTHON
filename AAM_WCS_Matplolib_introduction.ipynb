```
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import datetime as dt

link = "https://raw.githubusercontent.com/murpi/wilddata/master/quests/weather2019.csv"
df_weather = pd.read_csv(link).iloc[:,[0, 1, 2, -3]]

df_weather['DATE'] = pd.to_datetime(df_weather['DATE'])
```

```
# Complétez le code ci-dessous :

# Définition de la figure global
fig, ax = plt.subplots(figsize = (20,10)) #Le paramètre "figsize" permet de définir la largeur et la hauteur de la figure

# Définition du 1er graphique
ax1 = plt.subplot(242)
ax1.scatter(x = df_weather["DATE"], y = df_weather["MAX_TEMPERATURE_C"], c= 'yellow') # évolution des températures MAX en fct du temps
ax1.set_yticks(range(-5,50, 5)) # les bornes doivent être fixes! le nombre au milieu plus il est grand, moins les intervalles sont petites!
plt.ylabel("Temperature MAX")
ax1.set_title("Temperature MAX") # Affichage des titres sur l'axe "x"

# Définition du 2nd graphique
ax2 = plt.subplot(241) # le plot sur les "axe2" définie doit être sensiblement proportionelle(de une unité) au 1er graphe pour aligner sur la même ligne
ax2.scatter(x = df_weather["DATE"], y = df_weather["MIN_TEMPERATURE_C"], c= 'brown') # évolution des températures MIN en fct du temps
ax2.set_yticks(range(-5,50, 5))
plt.ylabel("Temperature MIN")
ax2.set_title("Temperature MIN") # Affichage des titres sur l'axe "x"

plt.show() # Affichage du plot
```
