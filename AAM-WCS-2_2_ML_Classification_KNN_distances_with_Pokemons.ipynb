```python
# Mise en place du calclule des distances entre k vosins des pokémons et optimisations de ces dernières

import pandas as pd # Biblothèque pandas
import math # Fonctions Mathématiques
import numpy as np # Biblothèque Numpy
import operator #Les "Opérateurs standards" en tant que fonctions pour utiliser ces dernières(fct)!

#TOPO sur le module 'operator'
#Le module operator fournit un ensemble de fonctions correspondant aux opérateurs natifs de Python. Par exemple, operator.add(x, y) correspond à l'expression x+y. 
#Les noms de la plupart de ces fonctions sont ceux utilisés par les méthodes spéciales, sans les doubles tirets bas.
#Pour assurer la rétrocompatibilité, la plupart de ces noms ont une variante avec les doubles tirets bas ;
#la forme simple est cependant à privilégier pour des raisons de clarté
# En général, les opérations logiques s'appliquent aussi à tous les objets et implémentent les tests de vérité, d'identité et les opérations booléennes :
#operator.not_(obj)
#operator.__not__(obj)

 
# Défintion de la fonction contenant le(s) test(s) d'entrainement/ échantillon test, caractéristiques et initialisation de la distance:
def calc_euclidean(train_data, test_data, train_point, test_point, features):
	total_dist = 0

	for feature in features:

		dist = math.pow(test_data.loc[test_point, feature] - train_data.loc[train_point, feature], 2)
		total_dist += dist

	# d = sqrt((a1-b1)^2 + (a2-b2)^2 + (a3-b3)^2 + (a4-b4)^2) #Distance eudienne entre 2 points est racine_carré(différence des-coordonnées de ces points: le tout au carré!)
	return math.sqrt(total_dist)

def det_nearest_neighbors(train_data, test_data, test_point, features, k):
	# Boucle sur tout et calcule la distance euclidienne en appelant l'utilisation des dictionnaires
	dist_dict = {}

	for train_point in train_data.index:
		# Définir l'index du point d'entrainement "train' plus la distance euclidienne
		dist_dict[train_point] = calc_euclidean(train_data, test_data, train_point, test_point, features)

	# Obtenez les k valeurs les plus élevées
	dict_top_k = sorted(dist_dict, key=dist_dict.get, reverse=True)[:k]

	return dict_top_k


def KNN(test_data, train_data, features, target_col_name, k):
	for test_point in test_data.index:
		# Trouver le nombre k de voisins les plus proches
		nearest_neighbors = det_nearest_neighbors(train_data, test_data, test_point, features, k)

	  # Convertir les index en classes
		nearest_neighbors = train_data[train_data.index.isin(nearest_neighbors)]

		# Compter, grouper par classe
		agg_class = nearest_neighbors.groupby(target_col_name).size()
		
		# L'index maximum (Max count :"idxmax()") est une nouvelle classe
		test_data.loc[test_point, 'classify'] = agg_class.idxmax()
	
	# Déterminer la précision (coefficient de proportionalité entre le coeff de classification sur celui du test) avec la variable "accuracy"
	len_classify_equals_target = len(test_data[test_data[target_col_name] == test_data['classify']].index) 
	len_test_data = len(test_data.index)
	accuracy = len_classify_equals_target / len_test_data
				
```

```python
# Utilisation 

import pandas as pd # Biblothèque pandas
import math # Fonctions Mathématiques
import numpy as np # Biblothèque Numpy
import operator #Les "Opérateurs standards" en tant que fonctions pour utiliser ces dernières(fct)!
 

# Importez la/le(s) function/ module/ bibliothèque/ package relatif à l'agorithme KNN & les bayésiens
import KNN as knn
import bayesian as bayes



def main():
	# prepare la Data
	# charger le jeu de données pokemon.
	data = pd.read_csv("./Pokemon.csv")
	# Determine le target
	target_col_name = 'Type 1'

	# Caractéristiques 
	# On décide de ne pas utiliser 'Légendaire' et 'Génération' car la variation est trop faible. 
	# Les classes comme 'Bug' n'ont pas de carte légendaire, ce qui entraîne une erreur : 'Divide by zero error'.
	features = ['Total', 'HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']

	
  # Séparation de l'échantillon du dataset test/training "train" (80/20)
	msk = np.random.rand(len(data)) < 0.80
	train_data = data[msk]
	test_data = data[~msk]

	# Ajouter une colonne "classify" pour classifier par leurs index
	test_data['classify'] = test_data.index

	# Determine k = n^0.5
	k = int(math.pow(len(data.index), 0.5))

	# Obtenir des résultats
	test_data_knn, accuracy_knn = knn.KNN(test_data, train_data, features, target_col_name, k)
	test_data_bayes, accuracy_bayes = bayes.gaussian_naive_bayes(test_data, train_data, features, target_col_name)

	# Imprimer la précisionj
	print('Accuracy KNN classifier: ' + str(accuracy_knn))
	print('Accuracy Gaussian naive bayes classifier: ' + str(accuracy_bayes))

if __name__ == '__main__':
    main()
```
